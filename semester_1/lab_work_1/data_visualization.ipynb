# Упражнение 5

```python
x = [10, 21, 28, 37, 43, 52, 61, 72, 80, 87, 95, 103, 111, 116, 123, 131]
y = [0.956, 0.948, 0.937, 0.932, 0.925, 0.917, 0.906, 0.908, 0.911, 0.912,
     0.925, 0.931, 0.958, 0.977, 1.005, 1.069]
z = arange(min(x), max(x), 0.01)

p1 = polyfit(x, y, deg = 1)
p2 = polyfit(x, y, deg = 2)
p3 = polyfit(x, y, deg = 3)
p_f_1 = poly1d(p1)
p_f_2 = poly1d(p2)
p_f_3 = poly1d(p3)

print("Первой степени:")
print(p_f_1)
print("\nВторой степени:")
print(p_f_2)
print("\nТретьей степени:")
print(p_f_3)

plt.figure(figsize = (13, 13))
plt.errorbar(x, y, yerr = 0.006, fmt = 'bo')
plt.plot(z, p_f_1(z), label = r'$p^1(x)$')
plt.plot(z, p_f_2(z), label = r'$p^2(x)$', c = 'g')
plt.plot(z, p_f_3(z), label = r'$p^3(x)$', c = 'r')

plt.legend(loc = 'best', fontsize = 20)
plt.grid()
plt.show()
```

# Упражнение 6

```python
a = 0
while a % 2 == 0 or a == 1 or a < 0:
    a = int(input("Введите нечетное a, не равное 1: a = "))
b = 0
while b <= 0 or b >= 1:
    b = float(input("Введите положительное число, меньшее 1: b = "))

xmin = float(input("Введите минимальный x: "))
xmax = float(input("Введите максимальный x: "))
step = float(input("Введите шаг счета: "))
x = arange(xmin, xmax, step)
n0 = int(100 / (xmax - xmin))
print("=================== " + str(n0) + " ===================")

j = 0
y = []
for x0 in x:
    y.append(0)

    for n in range(n0 + 1):
        y[j] += b**n * cos(a**n * pi * x0)
    
    j += 1

plt.figure(figsize = (15, 10))
plt.plot(x, y)

plt.show()
```
